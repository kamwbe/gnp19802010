```python
import numpy as np
import pandas as pd
import re
```

First I went to https://esa.un.org/unpd/wpp/Download/Standard/Population/ and downloaded 'Total Population both Sexes'.  I opened it in LibraOffice and saved it both as an .ods and a .cvs file to see which would be easier reading into a DataFrame.  I exported the file from LibraOffice as an html file and used the Pandas read_html to read it.

```python
worldPop = pd.read_html('worldpop19502015.html')
```

```python
# data read from and html file is loaded into a list of dataframes since there can be 
worldPopdf = worldPop[0]
worldPopdf.head(5)
```

```python
# from the above dump we can see that the column names are part of the data.  First
# we need to extract those column names and reset the df to have those column names
new_columns = worldPopdf.iloc[0,:]
worldPopdf.columns = new_columns
worldPopdf.head(5)
```

```python
# now we need to get rid of the first row which has the column names
# we do this by copying the df starting with row 2
worldPopdf = worldPopdf.iloc[1:] 
worldPopdf.head(3)
```

```python
worldPopdf.drop('Variant', axis=1, inplace=True)
```

```python
worldPopdf.drop('Notes', axis=1, inplace=True)
```

```python
worldPopdf.drop('Index', axis=1, inplace=True)
worldPopdf.head(3)
```

```python
# pat1 will look for numbers up to 999 999 999 999.  So the first re finds the number with spaces
# in it and the second re removes the spaces.  Finally we return the number as a float.
pat1=r'(\d+\s+\d*\s?\d*\s?\d*)'
re1=re.compile(pat1)
```

```python
# I wanted this to work over all the cells, even the Region... and Country code columns.
# The first if is for the Country code.  It should not match pat1 because it has no space
# (the pat1 has the \s+) and the the Region... column is alpha so it doesn't match.
# If it doesn't match then num == None and I just return x.
def pop_modifier(x):
    num = re1.search(x)
    if num == None:
        return x
    else:
        return float(re.sub('\s', '', num.group(), count=0, flags=0))
```

```python
# ok now look at the numbers in the year columns.  Instead of having an int or a float the thousands of the
# number has spaces between them.  So the df treats them as separate.  We need to convert this into one
# floating point number.
wPop = worldPopdf.applymap(pop_modifier)
```

```python
wPop.head(5)
```

```python
wPop.to_csv('clean_worldPop19502015.csv', line_terminator='\n', index=False, header=True)
```
