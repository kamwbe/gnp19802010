```python
# now we need to get rid of the first row which has the column names
# we do this by copying the df starting with row 2
worldPopdf = worldPopdf.iloc[1:] 
worldPopdf.head(3)
```

```python
# ok now look at the numbers in the year columns.  Instead of having an int or a float the thousands of the
# number has spaces between them.  So the df treats them as separate.  We need to convert this into one
# floating point number.
worldPopdf.loc[1]['1950']
```

```python
# pat1 will look for numbers up to 999 999 999 999.  So the first re finds the number with spaces
# in it and the second re removes the spaces.  Finally we return the number as a float.
pat1=r'(\d+\s?\d*\s?\d*\s?\d*)'
re1=re.compile(pat1)
```

```python
def pop_modifier(x):
    num = re1.search(x).group()
    num2 = re.sub('\s', '', num, count=0, flags=0)
    return float(num2)
```
